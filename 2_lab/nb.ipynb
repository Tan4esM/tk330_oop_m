- Запитуємо ChatGPT, 
> Для виведення піраміди за допомогою символів в мові програмування Python можна використовувати цикли та рядки. Ось приклад коду, який виведе піраміду зі зірочок:

```python
height = 5  # Встановлюємо висоту піраміди (кількість рядків)

for i in range(1, height + 1):
    # Виводимо пробіли перед зірочками (вирівнюємо по центру)
    print(" " * (height - i), end="")

    # Виводимо зірочки для поточного рядка
    print("*" * (2 * i - 1))
```

- Знайомимось з списками та кортеджі
> `List` - впорядкована колекція змінних типів даних;

> `Tuple` - впорядкована колекція незмінних типів даних;

```python
l = [1, 1.2, "текст", a]
ll = list(l)

print("Початкове значення списку:",l)

l.append("Це кінець") # Вставляємо в кінець
l.insert(1, "Це буде на індексі l") # Вставляємо на позицію l
l.pop(-2) # видаляємо 5

print("Після змін, список має вигляд:",l)

l.extend(ll)
print("Якщо зчепити два списки, буде:", l)
```

```python
t = (1, 1.2, "текст", a, 1)
tt = tuple(t)

print("Початкове значення кортеджу:",t)

print("Навіть використовуючи зчеплення, ми не модифікували наш tuple, а просто повернули нове значення:", t.__add__((1,)))

print("Кількість 1 буде:",t.count(1))
print("Знайдемо під яким індексом знаходиться число 1:",t.index(1))
```

- Знайомство з `Set`
> `Set` - набір унікальних, невпорядкованих даних

```python
s = {1, 1, "1", '1', 2, 'Кінець'}
print("Значення в Set будуть:", s)
print("Витягнути елемент індексу не можемо, однак ми можемо просто взяти якись елемент за допомогою:",s.pop())
s.add('1') # Додали вже існуючий елемент
print("Якщо хочемо працювати з елементами Set, то його перетворюють в list:", list(s), list(s)[1])
```

```python
a = "Ключ"
d = {1: "Це значення при ключі 1", "1": 1, a: "Це значення для ключа в змінній a", "a": "Це значення при ключі a"}

print("Введений словник буде мати вигляд:\n", d)

print("Щоб доступитись до якогось елемента, вказуємо його ключ",d[1],
      "\nЯкщо ми не знаємо, які існують ключі,",
      "\nто можна використати:", d.keys(),
      "\n\nМожна витягнути всі значення, перетворити в список і працювати елементами,",
       "\nяк з простими даними (візьмемо третій елемент з утвореного списку, який відповідає ключ 'Ключ'):", list(d.values())[3],
      "\n\nАбо можна було це зробити простіше і просто використати ключ a:",d["a"])
```

2. використовуємо метод стрічок `format`

```python
sf = "Це буде {} стрічка, де можна використати {} вставок".format("форматована", "декілька")
print(sf)
a = 1
b = 2
print("Значення a: {} та значеня b: {}".format(a , b))
# Можна вказувати, на які саме позиції вставляти наші вставки
print("Беремо останню: {2}, Беремо першу: {0}, та нарешті - ту що посередині: {1}".format("Перша позиція",
                                                                                           "Друга позиція",
                                                                                           "Третя позиція"))
print("Моє ім'я: {name}\nМоє хобі: {hobbi}\n{name} любить {hobbi}".format(name="Tatiana", hobbi="програмування"))
print("Значення змінної a: {var_a}, Не путати, що тут ми робимо позначку таку саму як змінна b: {b}".format(var_a=a, b=b))
```
