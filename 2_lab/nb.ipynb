```python
c = a + "|Це буде рядок зчеплення|" + b
d = a.__add__("|Це буде рядок зчеплення|".__add__(b) )

print("В обох випадках будемо мати:\n", c, "\n", d)

# Ці два записи рівноцінні
e = a * 4
f = a.__mul__(4)
print("Що буде при множенні стрічки на число?\n", e, "\n", f)
# Найпростіше, можна побавитися у вивід
print("~=~=~" * 25)
```

- Запитуємо ChatGPT, 
> Для виведення піраміди за допомогою символів в мові програмування Python можна використовувати цикли та рядки. Ось приклад коду, який виведе піраміду зі зірочок:

```python
height = 5  # Встановлюємо висоту піраміди (кількість рядків)

for i in range(1, height + 1):
    # Виводимо пробіли перед зірочками (вирівнюємо по центру)
    print(" " * (height - i), end="")

    # Виводимо зірочки для поточного рядка
    print("*" * (2 * i - 1))
```

- Числові типи даних

```python
a = 5
b = int(5)

print("Цілі числа автоматично будуть відноситися до класу:", type(a), type(b))
c = 5.5
d = float(a)

print("Якщо число є дробовим, то буде тип:", type(c), 
      "\nАбо можна явно перетворити", type(a), "в", type(d),
      "\nПісля перетворення число", a, "тепер буде", d)
```

```python
b += 0.1 # Це ще називають накопичуванням
print("В цьому випадку маємо явне перетворення: ", type(b), "\nРезультат (буде з похибкою):",b)

e = a + 0.1
f = 0.1 + a
print("\nТут все нормально працює:", type(e), "значення:", e,
      "\nТут також результат буде:", type(f), " значення:", f,)

e = a.__add__(0.1)
f = 0.1.__add__(a)
print("З використання методів класу додавання буде працювати тільки за певних умов.",
      "\nДля цілих типів даних можна додавати тільки цілі, інакше результат буде:", type(e), "значення:", e,
      "\nЯкщо ж цілі числа будем додавати до з плаваючою крапкою то буде: ", type(f), "згачення:", f,)
```

- Знайомимось з списками та кортеджі
> `List` - впорядкована колекція змінних типів даних;

> `Tuple` - впорядкована колекція незмінних типів даних;

```python
l = [1, 1.2, "текст", a]
ll = list(l)

print("Початкове значення списку:",l)

l.append("Це кінець") # Вставляємо в кінець
l.insert(1, "Це буде на індексі l") # Вставляємо на позицію l
l.pop(-2) # видаляємо 5

print("Після змін, список має вигляд:",l)

l.extend(ll)
print("Якщо зчепити два списки, буде:", l)
```

```python
t = (1, 1.2, "текст", a, 1)
tt = tuple(t)

print("Початкове значення кортеджу:",t)

print("Навіть використовуючи зчеплення, ми не модифікували наш tuple, а просто повернули нове значення:", t.__add__((1,)))

print("Кількість 1 буде:",t.count(1))
print("Знайдемо під яким індексом знаходиться число 1:",t.index(1))
```

- Знайомство з `Set`
> `Set` - набір унікальних, невпорядкованих даних

```python
s = {1, 1, "1", '1', 2, 'Кінець'}
print("Значення в Set будуть:", s)
print("Витягнути елемент індексу не можемо, однак ми можемо просто взяти якись елемент за допомогою:",s.pop())
s.add('1') # Додали вже існуючий елемент
print("Якщо хочемо працювати з елементами Set, то його перетворюють в list:", list(s), list(s)[1])
```
