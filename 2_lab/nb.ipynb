## Загальні поняття
> _Клас - це загальний темплейт який описує щось конкретне_;

> _Об'єкт - це окремий екзепляр класу зі своїми характеристиками/ параметрами_;

> _Тип даних - це характеристика набору даних_;

- вбудовані типи даних( текстові/стрічкові)

```python
print("Це є текстові дані, які будуть виведені")
a = "Ця змінна містить текстові дані"
print(a)
b = str("Це будуть текстові дані, через клас str")
print(b)

print("Рахуємо скільки разів зустрічаться буква е в об'єкті b: ", b.count("е"))
print("Рахуємо скільки разів зустрічаться буква і в об'єкті a: ", b.count("і"))
print("Рахуємо скільки разів зустрічаться буква і в простій стрічці: ", "Це буде стрічка без присвоєння змінній чи створення об'єкту".count("і"))
```
