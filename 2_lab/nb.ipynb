- Вбудовані типи даних( текстові/стрічкові)

```python
print("Це є текстові дані, які будуть виведені")
a = "Ця змінна містить текстові дані"
print(a)
b = str("Це будуть текстові дані, через клас str")
print(b)

print("Рахуємо скільки разів зустрічаться буква е в об'єкті b: ", b.count("е"))
print("Рахуємо скільки разів зустрічаться буква і в об'єкті a: ", b.count("і"))
print("Рахуємо скільки разів зустрічаться буква і в простій стрічці: ", "Це буде стрічка без присвоєння змінній чи створення об'єкту".count("і"))
```

- Контатенація (дjдавання/зчеплення) та множення стрічок

```python
c = a + "|Це буде рядок зчеплення|" + b
d = a.__add__("|Це буде рядок зчеплення|".__add__(b) )

print("В обох випадках будемо мати:\n", c, "\n", d)

# Ці два записи рівноцінні
e = a * 4
f = a.__mul__(4)
print("Що буде при множенні стрічки на число?\n", e, "\n", f)
# Найпростіше, можна побавитися у вивід
print("~=~=~" * 20)
```

- Запитуємо ChatGPT, 
> Для виведення піраміди за допомогою символів в мові програмування Python можна використовувати цикли та рядки. Ось приклад коду, який виведе піраміду зі зірочок:

```python
height = 5  # Встановлюємо висоту піраміди (кількість рядків)

for i in range(1, height + 1):
    # Виводимо пробіли перед зірочками (вирівнюємо по центру)
    print(" " * (height - i), end="")

    # Виводимо зірочки для поточного рядка
    print("*" * (2 * i - 1))
```
