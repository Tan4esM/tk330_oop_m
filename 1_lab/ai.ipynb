### Змінні у Python
Змінні у Python можна задавати локально (у межах функції або блоку коду) або глобально (на рівні всього скрипта). Ось приклади:

#### Локальні змінні:

Локальні змінні створюються та існують лише у межах функції або блоку коду, де вони були оголошені.

```python
def my_function() -> str: 
    # Це локальна змінна, доступна лише у межах функції.
    local_variable = "Це локальна змінна"
    print(local_variable)
    return local_variable

value_from_function = my_function()  # Виклик функції
#print(local_variable)  # Генерує помилку, бо змінна local_variable не існує за межами функцїі.

print(f"Функція поверне значення: {value_from_function}") # Якщо ми повернемо значення локальної змінної та перезапишемо її у іншу змінну, тоді ми зможемо користуватись її значенням!"
```

### Модифікована програма зі змінними та введенням значень з клавіатури
Ось модифікований код програми з використанням змінних та введенням значень з клавіатури:

```python
# Запитайте користувача про ім'я та збережіть його у змінній name
name = input("Будь ласка, введіть ваше ім'я: ")

# Виведіть привітання, використовуючи змінну name
print(f"Привіт, {name}!")

# Запитайте користувача про вік та збережіть його у змінній age
age = input("Скільки вам років? ")

# Виведіть вік користувача, використовуючи змінну age
print(f"Ваш вік: {age} років.")
```

У цьому коді ми використовуємо змінну name для збереження введеного користувачем імені та змінну age для збереження введеного віку. Функція input() використовується для отримання введення користувача з клавіатури. Команда print() використовує змінні для виведення інформації на екран.

Ця програма спершу запитає ім'я користувача, а потім його вік, і виведе відповіді на екрані.
